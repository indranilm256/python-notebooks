```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_stock_price(S0=100, T=1.0, N=1000, mu=0.1, sigma=0.2):
    dt = T / N
    t = np.linspace(0, T, N+1)
    dB = np.random.randn(N) * np.sqrt(dt)
    B = np.cumsum(dB)
    B = np.insert(B, 0, 0.0)

    S = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * B)
    return t, S

# Simulate and plot
t, S = simulate_stock_price()
plt.figure(figsize=(10, 4))
plt.plot(t, S)
plt.title("Simulated Stock Price using Geometric Brownian Motion")
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.grid(True)
plt.show()
```
