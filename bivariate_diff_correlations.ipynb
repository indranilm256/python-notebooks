```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Create a grid
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# Define multiple bivariate normal distributions with different correlation (rho)
distributions = {
    'Independent ( = 0)': {
        'mean': [0, 0],
        'cov': [[1, 0], [0, 1]]
    },
    'Positive Correlation ( = 0.8)': {
        'mean': [0, 0],
        'cov': [[1, 0.8], [0.8, 1]] 
    },
    'Negative Correlation ( = -0.8)': {
        'mean': [0, 0],
        'cov': [[1, -0.8], [-0.8, 1]]
    },
    'Highly Correlated ( = 0.99)': {
        'mean': [0, 0],
        'cov': [[1, 0.99], [0.99, 1]]
    }
}

# Plotting
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.ravel()

for i, (title, params) in enumerate(distributions.items()):
    rv = multivariate_normal(mean=params['mean'], cov=params['cov'])
    Z = rv.pdf(pos)
    axs[i].contourf(X, Y, Z, levels=30, cmap='coolwarm')
    axs[i].set_title(title)
    axs[i].set_xlabel("X")
    axs[i].set_ylabel("Y")
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```
