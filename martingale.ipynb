```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate a simple symmetric random walk (martingale)
np.random.seed(42)
n_steps = 100
n_walks = 5

walks = np.zeros((n_walks, n_steps + 1))

for i in range(n_walks):
    steps = np.random.choice([-1, 1], size=n_steps)  # Fair coin: -1 or +1
    walks[i, 1:] = np.cumsum(steps)

# Plot the walks
plt.figure(figsize=(10, 6))
for i in range(n_walks):
    plt.plot(walks[i], label=f"Walk {i+1}")
    
plt.axhline(0, color='black', linestyle='--', alpha=0.6)
plt.title("Simulated Random Walks (Martingales)")
plt.xlabel("Step")
plt.ylabel("Position")
plt.legend()
plt.grid(True)
plt.show()
```
