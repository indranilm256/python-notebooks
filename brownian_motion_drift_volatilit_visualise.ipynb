```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_brownian_motion(T=1.0, N=1000, mu=0.0, sigma=1.0):
    dt = T / N
    t = np.linspace(0, T, N+1)
    dB = np.sqrt(dt) * np.random.randn(N)
    B = np.cumsum(dB)
    B = np.insert(B, 0, 0.0)  # B_0 = 0

    X = mu * t + sigma * B  # Add drift and volatility
    return t, X

# Parameters to test
params = [
    {"mu": 0.0, "sigma": 1.0, "label": "Standard Brownian Motion"},
    {"mu": 1.0, "sigma": 1.0, "label": "Drift = 1.0, Volatility = 1.0"},
    {"mu": 0.5, "sigma": 2.0, "label": "Drift = 0.5, Volatility = 2.0"},
    {"mu": -0.5, "sigma": 0.5, "label": "Drift = -0.5, Volatility = 0.5"},
]

# Plot
plt.figure(figsize=(12, 6))
for p in params:
    t, X = simulate_brownian_motion(mu=p["mu"], sigma=p["sigma"])
    plt.plot(t, X, label=p["label"])

plt.title("Brownian Motion with Different Drift and Volatility")
plt.xlabel("Time")
plt.ylabel("X(t)")
plt.grid(True)
plt.legend()
plt.show()
```
