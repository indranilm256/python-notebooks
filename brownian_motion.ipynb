```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
T = 1.0            # total time
N = 1000           # number of steps
dt = T / N         # time step size
t = np.linspace(0, T, N+1)  # time grid

# Generate Brownian motion
dB = np.sqrt(dt) * np.random.randn(N)  # normal increments
B = np.zeros(N+1)
B[1:] = np.cumsum(dB)  # cumulative sum to get Brownian path

# Plotting
plt.figure(figsize=(10, 4))
plt.plot(t, B, label="Brownian Motion")
plt.title("1D Brownian Motion Simulation")
plt.xlabel("Time")
plt.ylabel("B(t)")
plt.grid(True)
plt.legend()
plt.show()
```
